## Conditional (If/Else)

Now that we know how to tell whether a comparison is true or false, we can build that into our code. "Conditional" statements allow us to do different things depending on the result of a comparison or Boolean variable, which we refer to as a condition. The logic looks like this:

* if a condition is true, then do something.

* if a conditions is false, do something else (or do nothing).

````python
if 1 > 0: # this statment checks the value, then IF TRUE it advances to the next line
    print("1 is greater than 0") # here where it prints the value
    print("test")
````

Key issue, an **`if`** statement must always be followed by **"`:`"** and then the next line or lines associated with the statement must be indicated by 4 spaces. ONLY 4 spaces, Jupyter will do this automatically....

````python
x = 5               # we can change this later and see what happens

if x > 6:
    print('x =', x)

print('Done!')
````

````python
x = 7

condition = x > 6

if condition:
    print('if branch')             # do if true
    print(condition)
else:
    print('else branch')           # do if false
    print(condition)
````

## `for` Loop

As with the conditional operation, this is a key skill. In particular, it allows for us to scale up the number of operations without having to constantly write code to perform each one. 

Suppose you have bunch of datasets. How to you read them in. Use a for loop. So you create a list of the names of the data sets, then you work through the list and read each one it. What the for loop allowed you to do is for a small change in your code you have scaled your operation by however much.

````python
namelist = ['Peter', 'Dave', 'Sarah', 'Robert']    # creates the list "namelist"
# below, the word "item" is arbitrary. End the line with a colon.

for item in namelist:    # goes through the items in the list one at a time
    print(item)    # indent this line exactly 4 spaces
# if there is code after this, we'd typically leave a blank line in-between
````

One thing to notice is the `item` which is a place holder. The specific name `item` has no attachment to anything, all it says is that as it works through each object in the list, its temporary name will be `item`. 

**Exercise**: Here, try to change `item` to `firstname` and see what happens.

Here is a numerical calculation...this is common in scientific computing...

````python
numlist = [4, -2, 5]
total = 0

for num in numlist:
    total = total + num

print(total)
````

Now here is a string. 

````python
word   = 'anything'
for letter in word:
    
    print(letter)
````

Note how general this is. The key issue is that as long as the for loop is over an "iteratable" object, then this this will work. Ask your self, what iterable objects do we know.

Now this example below combines the for loop and a conditional

````python
vowels = 'aeiouy'
word   = 'anything'

for letter in word:
    
    if letter in vowels:
        
        print(letter)
````

## List Comprehension

The basic idea is to create implicit loops on iterable objects like lists (note it need not be a list, just something we can iterate on). We refer to this as list comprehensions. It is useful in certain circumstances and shows up a lot in Python code. 

To understand list list comprehension, lets first remind our selves what a loop is doing

````python
namelist = ['Peter', 'Dave', 'Sarah', 'Robert']

for item in namelist:
    
    print(item)
````

A list comprehension gives us more compact syntax for the same thing:

````python
[print(item) for item in namelist];
````

Notices what this is doing:
- First there are the square brackets, this means we are creating a NEW LIST

- The first part within the brackes is the operation we want to perform on the old list. Here we want to print it.

- Then after the operation on the "item" we describe the for operation, the variable, and then the iterable object we want to grab from. 

- Essentialy what all this says is "do ``operation`` on ``item`` in the list"

Again, as with loops, the variable `item` can be any name we wish.  Replace `item` with your pet's name to see for yourself.

Lets see some examples:

````python
fruit = ['apple', 'banana', 'clementine'] # A list...

FRUIT = [item.upper() for item in fruit] 

print(FRUIT)
````

We can also apply conditions within list comprehensions

````python
fruit6 = [item for item in fruit if len(item)<=6]
````

## Linear Regression in Python

We will use the package `statsmodels` to run linear regressions.

````python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt # import the library for plotting
import seaborn as sns # import the library of beautiful plots
import statsmodels.api as sm # import library for linear regression
````

Before running the regression, we often first take a look at the data.

Let's load the data.

````python
stock_returns = pd.read_csv('returns.csv')
````

````python
stock_returns.head()
````

````python
# set figure parameters
params = {
   'figure.figsize': [6, 4],
   'xtick.labelsize': 10,
   'ytick.labelsize': 10
   }
plt.rcParams.update(params)

sns.set_context("paper")

fig, ax = plt.subplots()

stock_returns['GMA'] = stock_returns['RET_GE']-stock_returns['RET_APPLE']

sns.regplot(x='GMA', y='RET_GE',data=stock_returns)

ax.set_xlabel('the difference btw GE and Apple returns')
ax.set_ylabel('GE returns')

ax.set_title('Estimating the weights via the OLS',fontsize=12)

plt.tight_layout()
````

We use the OLS formula in `statsmodels` to build our model. The formula of the simple regression is

`Y ~ 1 + X`

````python
mod1 = sm.OLS.from_formula('RET_GE~1+GMA',data=stock_returns)
````

Then, we use `.fit()` to fit the data and save the regression results in `result1`.

````python
result1 = mod1.fit(cov_type='HC1',use_t=True)
````

Finally, we print the regression results.

````python
print(result1.summary())
````

The coefficient in front of `GMA` is the weight we should put on `Apple` if we want to have a minimum variance portfolio with `GE` and `Apple`.
